## 2.1 Change IMDB rating to categories
Create 3 different categories given the ratings:
- Good (7 to 10)
- Average (5 to 7)
- Bad (0 to 5)

```python
# define lables and categories
labels = ['Bad', 'Average','Good']
category = [0., 5., 7., 10.]

# create categories
moviesDF['Categories'] = pd.cut(moviesDF['IMDb'], labels=labels, bins=category, include_lowest=False)

moviesDF.head()
```

## 2.2 Clean NULL values
Change `NA` values in the column `age` to `Not Rated`

```python
moviesDF.fillna({'Age':'Not Rated'}, inplace = True)

moviesDF.head()
```

## 3.9 Create a word cloud visualization for the Genres.

```python
GenAN = ','.join(moviesDF['Genres'].dropna().tolist())

sns.set(rc={'figure.figsize':(11.7,8.27)})
wordcloud = WordCloud(background_color="white").generate(GenAN)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.title('Genres', size=24)
plt.show()
```

The most frequent and highest-rated movie genre is Drama, reflecting audience preferences. Comedy, Romance, Crime, and Sci-Fi are also popular. which indicates that viewers prefer some emotionally engaging stories with elements of humor, love, suspense, and futuristic concepts.

### 3.9a Analyse the frequency of the words then set a threshold for visualization
Remove too much frequent words and too less frequent words.

```python
# count the frequency of each genre

words = GenAN.split(',')
word_freq = pd.Series(words).value_counts()
word_freq
```

```python
# set the thredholds of frequencylower than 1500higher than 50 based on the frequencys

filtered_words = []

for word in words:
  if 50 <= word_freq[word] <= 1500:
    filtered_words.append(word)

filtered_gen_word = ' '.join(filtered_words)

wordcloud = WordCloud(background_color="white").generate(filtered_gen_word)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.title('Filtered Genres', size=24)
plt.show()
```

## RQ1: What is the distribution of age rating for movies on different platforms

```python
# exclude the Not Rated value in Age columns
df_Age = moviesDF.loc[moviesDF['Age'] != 'Not Rated'].groupby(by = 'Age')[['Netflix','Prime Video','Hulu','Disney+']].sum()

#sort the dataframe by age
row_order = ['all', '7+', '13+', '16+', '18+']
df_Age = df_Age.reindex(row_order)

# create the subplots
df_Age.plot(kind = 'barh', subplots = True, layout = (2, 2), sharex = True, sharey = True, legend = False)

plt.show()
```

**Anaysis:**

Prime Video has a significant advantage over its competitors in terms of movie copyrights, with Netflix following closely behind. However, Hulu and Disney+ have relatively fewer movie copyrights. This indicates that Prime Video has a broader range of movies available due to its extensive collection of copyrights.

When it comes to the distribution of ages, Prime Video, Netflix, and Hulu tend to focus more on adult-themed movies suitable for viewers aged 18 and above. Especially on Prime Video,  there are more than 1000 adults movies. On the other hand, Disney+ more focus on young audience like children, on which the mojority of movies are are suitable for young viewers of all ages. These differences in content indicates the different market segmentation positions and targeted audience of each platform.

Interestingly, all four platforms seem to have a limited number of movie resources for the 16+ teenage age group.  The reason may be the inherit film rating system setting lacking this age period. Or, it suggests that there is a potential gap in meeting the movie preferences of teenagers. To address this, each platform may need to invest more in developing and acquiring movies that resonate with this age group.

In summary, Prime Video's advantage in movie copyrights, the varying content strategies among Prime Video, Netflix, Hulu, and Disney+, and the need for more movie resources targeting the 16-18-year-old age group are key findings from the analysis. These insights highlight the importance of understanding audience preferences and catering to different age groups to stay competitive in the streaming industry.

## RQ2: What is developing trend of movies with different qualites in different plarforms after 2000?

```python
#filter the movies released after 2000
movies_2000 = moviesDF[moviesDF['Year'] >= 2000]

#create the pivot table
table1 = pd.pivot_table(movies_2000, values=['Netflix','Prime Video','Hulu','Disney+'], index=['Year'], columns=['Categories'], aggfunc='sum')

table1.tail(10)
```

**Anaysis:**


It this research question, I explored the trends in the number of movies with different qualities on each platforms after 2000.

It obviously is observed that film industry was developing rapidly before 2017, but experienced a sharp decline by 2020. This could be caused by Covid-19, which led to a significant decrease in the global movie releases.

There are significant increases in the number of movies on Netflix and Hulu after 2015, while Prime Video seemed to have started its development earlier, with a notable increase after 2007. Disney's data is limited, showing unusual fluctuations, but it can still be observed that movies remained relatively less than other platforms.

In terms of movie quality, the majority of movies on each platform fall into the average level. However, A more uneven distribution of quality happened on Prime Video, with a higher number of movies categorized as Bad" compared to Good. On contrast, Netflix shows the opposite trend, with a higher number of Good" movies than Bad." This may indicate that Netflix is more strict in ensuring movie quality and prefer to purchase the copyrights of high-quality films. As for Disney+ and Hulu, the distribution of good and bad movies appears to be more balanced.
